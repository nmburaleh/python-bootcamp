## Creating a Shopping Cart

### Final Design

As we’ve introduced functions this week, the final design will be based around the logic
of our program’s actions. Functions perform a specific task, which is usually an action.
For our shopping cart program, the actions that we need to consider are the tasks of
adding, removing, clearing, and showing the items within the cart. The logical design
will look like Figure 5-5.

We’ll be sure to have one main function that will contain the loop and handle user
input.

```python
from IPython.display import Image
Image("../../Pictures/figure_5_5.png")
```

```python
# import necessary functions
from IPython.display import clear_output

# global list variable representing user shopping cart
cart = []

# function to add items to cart
def add_item():
    user_item = input("What would you like to add? ").casefold() # 4
    clear_output()
    
    cart.append(user_item)
    print(f"{user_item} has been added")
    
# function to remove items from cart
def remove_item():
    user_item = input("What would you like to remove? ").casefold() # 4
    clear_output()
    
    try:
        cart.remove(user_item) # 1
        print(f"{user_item} has been removed.")
    except NameError: # 1
        print("Sorry that item is not the cart")
        
# function to show items in cart
def show_cart():
    clear_output()
    if cart:
        print("Here are the item(s) in your cart: ")
        for item in cart:
            print(f"-  {item}")
    else:
        print("Your cart is empty.")
        
# function to clear items from cart
def clear_cart():
    clear_output()
    cart.clear()
    print("Your cart is empty.")
    
options = {'add': add_item,
           'remove': remove_item,
           'show': show_cart,
           'clear': clear_cart} # 2

# "quit/add/remove/show/clear
def main():
    done = False
    while not done:
        user_ans = input("quit/add/remove/show/clear: ").casefold()
        
        # terminal condition
        if user_ans == "quit":
            print("Thank you for using the program")
            show_cart()
            done = True
            break
        
        # primary business logic
        if user_ans in options.keys(): # 3
            options[user_ans]()
        else:
            print("Incorrect input...try again!")
            
if __name__ == "__main__":
    main()
```

### Summary of Changes I've made from Milliken:

1. Explicitly stated the class of error that occurs when user inputs a value to remove from cart.
2. Created a mapping of user inputs to the non-terminal options (add, remove etc)
3. Made all non-terminal options parameterless to make the primary business logic concise and generalised.
4. Placed additional user input requirements within the scope of the appropriate functions.
